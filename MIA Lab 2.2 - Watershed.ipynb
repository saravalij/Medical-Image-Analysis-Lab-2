Importing modules and files' functions, and reading the images
</div></center>

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import matplotlib
from segmentation_f import *
```

```python
img = io.imread('Brain.tif',as_gray=True)
img = img/np.max(img)
img_c = io.imread('Chest.tif',as_gray=True)
img_c = img_c/np.max(img_c)
```


<center><div class="alert alert-block alert-warning">
Segmenting an image of the brain
</div></center>

```python
matplotlib.use('QT5Agg')
grad,im_image,combined = WatershedExerciseP2(img)
```

```python
%matplotlib inline
plot_img((grad,im_image,combined), ('Gradient Watershed', 'Imimposemin Watershed','Separate segmented regions'), cmap= 'jet')
```

- As we can see, the different colours indicate different things, the blue colour corresponds to the **maximums**, which in this case are the edges (foreground), and the red colour the **minimums** (background). These colours change progressively according to their intensities. In the gradient image (gradient watershed), the watershed algorithm has not been able to segment the image, so the colours are progressively ordered. However, in the second image (imimposemin watershed) we can see how the image is segmented a little and we can see the **edge of the brain**. 
-  In the gradient image alone the different segmentations are **not well distinguished**, if we use different seeds that are going to be present in our binary mask and apply the imimposemin function we get a better segmentation. In the third image (imimposemin watershed combined), we have tried to draw a **line** that separates our **segmented regions**, as we can see we are left with only the frame of the brain. 
- We use the **gradient image** as **input** because the edges of our image are marked on it and this is what we are going to take into consideration. If we used the img image instead, the result would not be the same as the edges would not be well defined. 
- One of the main problems of the watershed is the **over-segmentation**. To improve the quality of the image and solve this problem we can apply different **seeds** as we have done above. Other methods that can be used to obtain an improvement are; to apply a **smoothing filter** before applying the algorithm to reduce the noise in the image, or to use a method similar to split and merge and **join neighbouring regions** with similar characteristics.
