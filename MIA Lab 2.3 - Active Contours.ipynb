**Active models** have been widely used in biomedical **image segmentation**, these models provide an effective framework for object segmentation. In cases where the image contains artifacts or noise, it is difficult to extract objects from the images with high accuracy, so it is useful to have prior knowledge of the shape of the object to be segmented.  

The **goal** of the snake’s algorithm is to find a **contour** that **best approximate** to the **boundaries** of the object. It will consist on calculating internal and external forces. First, the snake is placed close to our frame of interest. The snake will be attracted by various forces that control the location and shape of the snake within the image, this process is iterative.

It is based on **minimizing an energy function** that contains several terms:   

<img src= https://wikimedia.org/api/rest_v1/media/math/render/svg/2e93dd59760c6d5d51826b30406b3bcd854a3862>

- The **E_external** is a combination of E_image and E_con. 
- **E_con**: restrictive forces.
- **E_image** is a function of the characteristics of the image, attracts contour toward the closest image edge (becomes very small when the snake points get close to an edge).
- **E_internal** is composed of the continuity and the smoothness of the contour. 

There are different energy settings **α**, **β** and **ϒ** that can be varied to change the parameters and select different regions.

Importing modules and files' functions, and reading the image
</div></center>

```python
import matplotlib.pyplot as plt
from skimage import io
from segmentation_f import activecontours
```

```python
img =io.imread('Brain.tif',as_gray=True)
```

<center><div class="alert alert-block alert-warning">
Segmenting the whole brain
</div></center>

We will use the following parameters:
    
- r = 200 + 210*np.sin(s)
- c = 170 + 160*np.cos(s)
- α = 0.005
- β = 50
- ϒ = 0.030

```python
snake, init = activecontours(img,r=[200,210],c=[170,160],act_param=[0.005,50,0.030]) #skull
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(img, cmap=plt.cm.gray)
ax.plot(init[:, 1], init[:, 0], '--r', lw=3)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, img.shape[1], img.shape[0], 0])

plt.show()
```

<center><div class="alert alert-block alert-warning">
Segmenting just two ventricles
</div></center>

To segment the two ventricles, we will modify the parameters as follows:

- s = np.linspace(0, 2*np.pi, 400)
- r = 140 + 40*np.sin(s)
- c = 175 + 50*np.cos(s)
- α = 0.015
- β = 50
- ϒ = 0.1

```python
snake, init = activecontours(img,r=[140,40],c=[175,50],act_param=[0.015,50,0.1]) #ventricles
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(img, cmap=plt.cm.gray)
ax.plot(init[:, 1], init[:, 0], '--r', lw=3)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, img.shape[1], img.shape[0], 0])

plt.show()
```
